```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn import metrics  
```

```python
data ="D:\Hours and Scores.csv"
sh_data = pd.read_csv(data)
print("Data imported successfully") 

```

```python
sh_data.head(10)


```

```python
sh_data.isnull()

```

```python
sh_data.notnull()
```

```python
sh_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours against Percentage of scores')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()

```

```python
X = sh_data.iloc[:, :-1].values  
y = sh_data.iloc[:, 1].values


```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")

```

```python
line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y)
plt.plot(X, line);
plt.title("Regression Line")
plt.show()
```

```python
print(X_test) 
y_pred = regressor.predict(X_test) 
```

```python
HS_data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
print(HS_data)
```

```python
hours=9.25
hours=np.array(hours).reshape(1, -1)
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))

```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred))
```
